## **Análisis de la vulnerabilidad del sistema de salud en México: Cobertura y crecimiento poblacional**

Con el objetivo de evaluar el estado del sistema salud a nivel nacional y el nivel de vulnerabilidad de las entidades federativas mexicanas ante un posible colapso del sistema de salud como resultado del cambio demografico y una presión creciente sobre los recursos del sistema de salud, se ha realizado un análisis cuantitativo basado en datos históricos (2002–2022) de crecimiento poblacional, cobertura médica (camas y médicos por cada 1,000 habitantes y población afiliada),el gasto público en salud y presión relativa sobre los recursos disponibles.



### **Variables**:


*  ***Médicos por cada 1000 habitantes***

    -Tipo de Recurso: Recurso Humano

    -Definición: Total de médicos activos en el sector salud por cada 1000
    habitantes en una área geográfica específica(país, ciudad, municipio, localidad, etc.)

    -Propósito: Identificar si la densidad de médicos/as de una región es suficiente para promover, mantener y recuperar la salud de su población. Además, es un indicador que permite monitorear la correcta distribución de recursos humanos y económicos para garantizar el Acceso Universal a la Salud y la Cobertura Universal de Salud

    https://opendata.paho.org/es/indicadores-basicos/fichas-tecnicas


 *  ***Camas censables por cada 1000 habitantes***

    -Tipo de Recurso: Recurso material

    -Definición: Total de camas funcionales en área de hospitalización en el sector salud por cada 1000 habitantes en una área geográfica específica(país, ciudad, municipio, localidad, etc.). l

    -Propósito: Valorar si la disponibilidad de camas censables es sufiente para atender a una población en un lugar y periodo determinado, así como también determinar el porcentaje de población cubierta  con base en este recurso.

    https://salud.gob.mx/unidades/evaluacion/indicadores/notas/indicador26.pdf

```python
# La siguiente función genera una tabla por estado y año con las siguientes variables:
# Estado, población, Camas Censables, Camas censables / 1000 habitantes, Médicos Totales, Médicos Totales/1000 habitantes
def Union_DataFrames(dfc,dfm,Estados):
  UniondfC=[]
  UniondfM=[]
  Tabla_Med_Cms=[]

  años1=2002
  años2=2002

#Para el Total de camas censables por estado
  for i in dfc:

     #Se Eliminan  las columnas que no son útiles
       i.drop(['State','Resources Categories','Resources Categories ID','State ID','Total'],axis=1,inplace=True)

      # Se Agrega el año a cada columna como sufijo y se renombran las columnas de interés
       i.rename(columns={"Population":"Población","Per 1000 inhabitants": "Camas/1000 habitantes"}, inplace=True)
       i=i.add_suffix(f'_{años1}')


       años1=años1+1
       UniondfC.append(i)
#Para el Total de médicos generales, especialistas y odontólogos por estado
  for j in dfm:
     #Se Elimina las columnas que no son útiles

     j.drop(['State','Population','Resources Categories','Resources Categories ID','State ID','Total'],axis=1,inplace=True)
     j.rename(columns={"Per 1000 inhabitants": "Médicos/1000 habitantes"}, inplace=True)
     j=j.add_suffix(f'_{años2}')
     años2=años2+1

     UniondfM.append(j)
  #  Se unen las columnas de interés por estado y año
  for k in range(len(UniondfC)):
    Tabla_Union=UniondfC[k].join(UniondfM[k])
    Tabla_Med_Cms.append(Tabla_Union)
  #Concatena los DataFrames
    Tabla_Final=pd.concat(Tabla_Med_Cms, axis=1)

    Tabla_Final.insert(0, 'Estado', Estados)



  return Tabla_Final
```

```python
# Tasas de crecimiento compuesto,TCAC

# Función para calcular la tasa de crecimiento compuesta, de 2002 a 2022. La tasa se expresa de manera
# porcentual.
def TCAC(df,columna_base,añoI,añoF):
  #df:DataFrame
  #columna_base: Columna/as a analizar
  #años: rango de años. Debe corresponder con los años presentes en el DataFrame.
  #Tasa de crecimiento compuesto,TCAC= [(Valor actual-Valor anterior)**(1/n)]-1
  # con n:= periodo de años

    periodo=añoF-añoI

    tasas_crecimiento = {}
    for i in range(len(df)):
        estado = df.iloc[i]['Estado']
        crecimiento_estado = []
        Valor_I=f'{columna_base}_{añoI}'
        Valor_F= f'{columna_base}_{añoF}'
        # Validación de columnas: Existencia y no nulidad
        if Valor_I in df.columns and Valor_F in df.columns:
                valor_anterior = df.iloc[i][Valor_I]
                valor_actual = df.iloc[i][Valor_F]
                # Transformación de variables a tipo numérico y manejo de posibles errores
                try:
                    valor_anterior = pd.to_numeric(valor_anterior, errors='coerce')
                    valor_actual = pd.to_numeric(valor_actual, errors='coerce')
                except:
                    crecimiento_estado.append(np.nan) # Si no se puede convertir, usar NaN
                    continue

                if pd.isna(valor_anterior) or pd.isna(valor_actual):
                    crecimiento_estado.append(np.nan)
                elif valor_anterior == 0:
                    # Evitar división por cero, considerar crecimiento infinito o 0
                    crecimiento_estado.append(np.nan if valor_actual > 0 else 0)
                else:
                    tasa =(valor_actual/valor_anterior)
                    tasa=tasa**(1/periodo)
                    tasa=tasa-1


                    crecimiento_estado.append(tasa)
        else:
                crecimiento_estado.append(np.nan) # Si la columna no existe, usar NaN

        tasas_crecimiento[estado] = crecimiento_estado

    # Se crea un nuevo DataFrame para las tasas de crecimiento
    años_tasa = [f'TCAC {columna_base},{añoI}-{añoF}']
    df_tasas = pd.DataFrame.from_dict(tasas_crecimiento, orient='index', columns=años_tasa)
    df_tasas.insert(0, 'Estado', df_tasas.index)
    df_tasas.reset_index(drop=True, inplace=True)


    return df_tasas






```

```python
# Cálculo de las Tasa de Crecimiento Anual Compuesto
TCAC_POB=TCAC(His_CmsYMeds,'Población',2002,2022)
TCAC_ICms=TCAC(His_CmsYMeds,'Camas/1000 habitantes',2002,2022)
TCAC_IMeds=TCAC(His_CmsYMeds,'Médicos/1000 habitantes',2002,2022)

```

## **Normalización de variables**


```python
from sklearn.preprocessing import MinMaxScaler
```

```python
# Se define el escalador
Escalador=MinMaxScaler(feature_range=(-1,1))
```

```python

# Se ajusta y transforma el escalador a TCAC_ICM
TCACs_ICM_Norm = Escalador.fit_transform(TCAC_ICM[['TCAC Camas/1000 habitantes,2002-2022', 'TCAC Médicos/1000 habitantes,2002-2022', 'TCAC Población,2002-2022']])

# Se crea un nuevo DataFrame con los datos escalados

TCAC_ICM_Norm_df = pd.DataFrame(TCACs_ICM_Norm,columns=['TCAC_Norm Camas/1000 habitantes,2002-2022', 'TCAC_Norm Médicos/1000 habitantes,2002-2022', 'TCAC_Norm Población,2002-2022'])
TCAC_ICM_Norm_df.insert(0, 'Estado', Estados)

print('Normalización de las Tasas de Crecimiento Anual Compuesto')
TCAC_ICM_Norm_df.head(10)
```

Bajo la suposición de que las razones médicos/1000 habitantes y camas/1000 habitantes, son recursos indispensables y complementarios para garantizar una atención médica efectiva, se definen los siguientes índices con el objetivo de generar un cluster para determinar el nivel de vulnerabilidad del sistema de salud por estado


### **Índice de Cobertura Médica(ICM)**
*  **Buena cobertura**: La tasa de crecimiento del recurso es mayor o igual que la tasa de crecimiento de la población -----> **ICM posiivo**

* **Cobertura deficiente**: La tasa de crecimiento del recurso es menor que la tasa de crecimiento de la población-----> **ICM negativo**

#### **Fórmula**

**ICM**=TCAC_Norm Médicos/1000 habitantes - TCAC_Norm Población

### **Índice de Cobertura Hospitalaria(ICH)**
*  **Buena cobertura**: La tasa de crecimiento del recurso es mayor o igual que la tasa de crecimiento de la población -----> **ICH posiivo**

* **Cobertura deficiente**: La tasa de crecimiento del recurso es menor que la tasa de crecimiento de la población-----> **ICH negativo **

#### **Fórmula**

**ICH**=TCAC_Norm Camas/1000 habitantes- TCAC_Norm Población



```python

# Índice de Cobertura

def Cobertura(df):
  #Se calculan los índices por estado

  ICM=df['TCAC_Norm Médicos/1000 habitantes,2002-2022']-df['TCAC_Norm Población,2002-2022']
  ICH=df['TCAC_Norm Camas/1000 habitantes,2002-2022']-df['TCAC_Norm Población,2002-2022']
  indices=pd.DataFrame(list(zip(ICM,ICH)),columns=['ICM','ICH'])
  indices.insert(0, 'Estado', Estados)



  return indices

```

```python
Indice_Cobertura=pd.DataFrame(Cobertura(TCAC_ICM_Norm_df))

print('Índice de Cobertura por estado, periodo 2002-2022')
Indice_Cobertura.head(10)
```

```python
# Normalización de los índices ICM e ICH
Indice_Cobertura_Norm = Escalador.fit_transform(Indice_Cobertura[['ICM', 'ICH']])
Indice_Cobertura_Norm_df = pd.DataFrame(Indice_Cobertura_Norm, columns=['ICM_Norm', 'ICH_Norm'])
Indice_Cobertura_Norm_df.insert(0, 'Estado', Estados)
```

```python
Indice_Cobertura_Norm_df.head(10)
```

# **Clusters**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Determinación del número de clusters mediante el uso del método del codo
inertia = []
silhouette = []
K = range(2, 9)  # evaluamos de a 9 clusters

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(Indice_Cobertura_Norm_df[['ICH_Norm', 'ICM_Norm']])
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(Indice_Cobertura_Norm_df[['ICH_Norm', 'ICM_Norm']], kmeans.labels_))
```

```python
# Gráfico del método del codo

Figure1=plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(K, inertia, marker='<')
plt.title('Método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('Inercia')

# Gráfico del índice de silueta
plt.subplot(1, 2, 2)
plt.plot(K, silhouette, marker='o', color='orange')
plt.title('Índice de silueta')
plt.xlabel('Número de clusters')
plt.ylabel('Score')

plt.tight_layout()
plt.show()
Figure1.savefig('Método_Codo_Silhouette.png')
```

* Menor valor inercial, mayor compactación.
* Índice de silueta: Si el valor del índice es  1, la muestras están lejos de clusters vecinos.Si el valor del índice es -1, indica una asignación de cluster erronea de la muestra. Los valores cercanos a 0 indican conglomerados superpuestos.

Bajo estos supuestos, se define que el número de clusters óptimos para este análisis es k=3

```python
k_optimo = 3
kmeans = KMeans(n_clusters=k_optimo, random_state=42)
Indice_Cobertura_Norm_df['cluster'] = kmeans.fit_predict(Indice_Cobertura_Norm_df[['ICH_Norm', 'ICM_Norm']])
```

```python
# Determinación de centroides
Centroides = kmeans.cluster_centers_
print(Centroides)
```

```python
Figure2=plt.figure(figsize=(8, 6))
sns.scatterplot(data=Indice_Cobertura_Norm_df, x='ICH_Norm', y='ICM_Norm', hue='cluster', palette='tab10')
plt.scatter(Centroides[:, 0], Centroides[:, 1], marker='X', color='red', s=50, label='Centroides')
plt.legend(title='Cluster')
plt.title('Clustering por ICH_Norm e ICM_Norm')
plt.xlabel('Índice de Cobertura Hospitalaria (ICH) Normalizado')
plt.ylabel('Índice de Cobertura Médica (ICM) Normalizado')

plt.grid(True)
plt.show()
Figure2.savefig('Clustering_ICH_ICM.png')
```
